```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
```

```python
t = np.arange(0.0,10.0,1/50.0)
```

```python
f1 = 3./4 * np.sin(1 * 2*math.pi*0.5*t)
```

```python
plt.plot(t,f1)
plt.grid()
plt.show()
```

```python
f2 = 1./4 * np.sin(3 * 2*math.pi*0.5*t)
```

```python
plt.plot(t,f2)
plt.grid()
plt.show()
```

```python
plt.plot(t,f1)
plt.plot(t,f2)
plt.grid()
plt.show()
```

```python
f_sum = f1 + f2
```

```python
plt.plot(t,f_sum)
plt.grid()
plt.show()
```

```python
f1_e = 3./4 * 1./2j * ( math.e**(1j*( 1 * 2*math.pi*0.5*t)) - 
                        math.e**(1j*(-1 * 2*math.pi*0.5*t)) )
```

```python
plt.plot(t,f1_e.real)
plt.grid()
plt.show()
```

```python
f2_e = 1./4 * 1./2j * ( math.e**(1j*( 1 * 3*2*math.pi*0.5*t)) - 
                        math.e**(1j*(-1 * 3*2*math.pi*0.5*t)) )
```

```python
plt.plot(t,f2_e.real)
plt.grid()
plt.show()
```

```python
plt.plot(t,f1_e.real)
plt.plot(t,f2_e.real)
plt.grid()
plt.show()
```

```python
f_sum_e = f1_e + f2_e
```

```python
plt.plot(t,f_sum_e.real)
plt.grid()
plt.show()
```
