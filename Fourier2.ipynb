```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
```

```python
t = np.arange(0.0,10.0,1/50.0)
```

```python
f1 = 3./4 * np.sin(1 * 2*math.pi*0.5*t)
```

Как рисовать описано кратенько вот тут

```python
plt.plot(t,f1)
plt.grid()
plt.show()
```

```python
f2 = 1./4 * np.sin(3 * 2*math.pi*0.5*t)
```

```python
plt.plot(t,f2)
plt.grid()
plt.show()
```

```python
plt.plot(t,f1)
plt.plot(t,f2)
plt.grid()
plt.show()
```

```python
f_sum = f1 + f2
```

```python
plt.plot(t,f_sum)
plt.grid()
plt.show()
```

```python
f1_e = 3./4 * 1./2j * ( math.e**(1j*( 1 * 2*math.pi*0.5*t))  
                       -math.e**(1j*(-1 * 2*math.pi*0.5*t)) )
```

```python
plt.plot(t,f1_e.real)
plt.grid()
plt.show()
```

```python
f2_e = 1./4 * 1./2j * ( math.e**(1j*( 1 * 3*2*math.pi*0.5*t))  
                       -math.e**(1j*(-1 * 3*2*math.pi*0.5*t)) )
```

```python
plt.plot(t,f2_e.real)
plt.grid()
plt.show()
```

```python
plt.plot(t,f1_e.real)
plt.plot(t,f2_e.real)
plt.grid()
plt.show()
```

```python
f_sum_e = f1_e + f2_e
```

```python
plt.plot(t,f_sum_e.real)
plt.grid()
plt.show()
```

## Покажем в 3D что по Re оси у нас sin, а по Im оси - cos.

```python
#здесь f_re это та же сумма синусов, а f_im сумма косинусов с той же частотой
f_re = 3./4 * np.sin(1 * 2*math.pi*0.5*t) + 1./4 * np.sin(3 * 2*math.pi*0.5*t)
f_im = 3./4 * np.cos(1 * 2*math.pi*0.5*t) + 1./4 * np.cos(3 * 2*math.pi*0.5*t)
```

```python
#рисуем в 3D!
from mpl_toolkits.mplot3d import axes3d 
```

Ничего сложного в коде для рисования в 3D нету,

можно прочитать https://www.geeksforgeeks.org/graph-plotting-python-set-2/

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.scatter(t, f_im,f_re)
ax1.set_xlabel('t') 
ax1.set_ylabel('f_im') 
ax1.set_zlabel('f_re')

ax2 = fig.add_subplot(1,2,2)
ax2.plot(t,f_sum)
ax2.set_xlabel('t')
ax2.set_ylabel('f_sum = f_re')
plt.show()
```

## Собственно, Преобразование Фурье

```python
from scipy.fftpack import fft, ifft
```

```python
#intital signal
discrete_signal = f_sum
```

```python
#Discrete Fourier Transform -> Spectrum
magnitudes_of_frequencies = fft(discrete_signal)
# один элемент здесь это комплексное число:
magnitudes_of_frequencies[1]
```

```python
frequency = np.arange(0,len(magnitudes_of_frequencies))
```

```python
plt.bar(frequency[0:250],magnitudes_of_frequencies.real[0:250])
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,1,1, projection='3d')
ax1.scatter(frequency[0:50], magnitudes_of_frequencies.imag[0:50],magnitudes_of_frequencies.real[0:50])
ax1.set_xlabel('freq') 
ax1.set_ylabel('imag') 
ax1.set_zlabel('real')
plt.show()
```
