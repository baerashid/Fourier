```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
from scipy.fftpack import fft, ifft
```

```python
#intital signal
x = np.array([2.0, 4.0, -1.0, 6.0])
```

```python
#Descrete Fourier Transform -> Spectrum
y = fft(x)
```

```python
for elem in y:
    print (elem)
```

```python
#Inverse Descrete Fourier Transform -> Original signal
yinv = ifft(y)
```

```python
for elem in yinv:
    print (elem)
```

Consider a sinusoidal signal f_t that is a function of time t with frequency components of 15 Hz and 20 Hz. Use a time vector sampled in increments of 1/50 of a second over a period of 10 seconds.

```python
t = np.arange(0.,1.,1/50.)
f_t = np.sin(2*math.pi*15*t) + np.sin(2*math.pi*30*t)
```

```python
plt.plot(t, f_t,marker='o') 
plt.grid()
plt.show()
```

Compute the Fourier transform of the signal, and create the vector f that corresponds to the signal's sampling in frequency space. 

```python
F = fft(f_t)
frequency = np.arange(0,50)
```

```python
plt.bar(frequency, abs(F))
```
